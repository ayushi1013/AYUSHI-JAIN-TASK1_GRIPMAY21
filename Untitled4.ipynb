# Predict the house sales in King County,Washington State, USA


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
house_data= pd.read_csv("kc_house_data.csv",parse_dates = ['date'])
house_data.head()
```

# Data Preprocessing


```python
last_column = house_data.pop('price')
house_data.insert(20, 'price', last_column)
```

**Preprocessing Date Column**

```python
house_data['date']= house_data['date'].astype(str)
house_data['year']=house_data['date'].apply(lambda x: x.split("-")[0])
house_data['month']=house_data['date'].apply(lambda x: x.split("-")[1])
house_data['date']=house_data['date'].apply(lambda x: x.split("-")[2])
```

```python
house_data['date'] = house_data['date'].astype('int')
house_data['month'] = house_data['month'].astype('int')
house_data['year'] = house_data['year'].astype('int')
```

```python
last_column = house_data.pop('month')
house_data.insert(2, 'month', last_column)
```

```python
last_column = house_data.pop('year')
house_data.insert(3, 'year', last_column)
```

```python
house_data.head()
```

# **Correlation Plot**

```python
f = plt.figure(figsize=(20, 15))
corrMatrix = house_data.corr()
sn.heatmap(corrMatrix, annot=True)
plt.show()
```

# Defining features

```python
x=house_data.drop(columns=['id','price'])
```

```python
x=x.drop(columns=['sqft_living15','sqft_above','zipcode'])
```

```python
x=x.drop(columns=['date','year'])
```

# Normalizing Data

```python
from sklearn.preprocessing import MinMaxScaler
nd=MinMaxScaler()
X=pd.DataFrame(nd.fit_transform(x))
X.columns=x.columns
X.index=x.index
X.head()
```

```python
y=pd.DataFrame(house_data['price'])
```

```python
Y=pd.DataFrame(nd.fit_transform(y))
Y.columns=y.columns
Y.index=y.index
Y.head()
```

# Fitting model

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size=0.3) 
```

```python
lr_model=LinearRegression()
lr_model.fit(X_train,y_train)
```

# Finding MAE,Score

```python
lr_value_predict=lr_model.predict(X_test)
lr_val_mae=mean_absolute_error(lr_value_predict,y_test)
lr_val_mae
```

```python
lr_model.score(X_train,y_train)
```

```python
lr_model.score(X_test,y_test)
```
